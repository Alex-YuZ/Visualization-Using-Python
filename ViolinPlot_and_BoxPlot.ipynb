### Example 1. Data Distribution in Efficiencies by Car Types Using `sns.violinplot()`

**Preparatory Step 1: Self-define a categorical data type.**

```python
# Types of sedan cars
sedan_classes = ['Minicompact Cars', 'Subcompact Cars', 'Compact Cars', 'Midsize Cars', 'Large Cars']

# Returns the types for sedan_classes with the categories and orderedness
# Refer - https://pandas.pydata.org/pandas-docs/version/0.23.4/generated/pandas.api.types.CategoricalDtype.html
vclasses = pd.api.types.CategoricalDtype(ordered=True, categories=sedan_classes)

# Use pandas.astype() to convert the "VClass" column from a plain object type into an ordered categorical type 
fuel_econ['VClass'] = fuel_econ['VClass'].astype(vclasses);
```

**Preparatory Step 2: Uniform the color to reduce unnecessary encoding.**

```python
# Uniform the color to reduce unnecessary encoding
base_color = sns.color_palette()[0]
```

**Step 3: Plotting with different `inner` settings:**

inner可选参数的取值： {“box”, “quartile”, “point”, “stick”, None}

```python
plt.figure(figsize=(23, 6))
inners_dict = {1: 'box', 2: 'quartile', 3: None}

for key, inner in inners_dict.items():
    plt.subplot(1, 3, key)
    sns.violinplot(data=fuel_econ, x='VClass', y='comb', color=base_color, inner=inner)
    plt.title("inner='{}'".format(inner))
    plt.xticks(rotation=15);
    
```

**Variations: Horizontal Violin Plot**

```python
sns.violinplot(data=fuel_econ, y='VClass', x='comb', color=base_color, inner='box')
plt.xticks(rotation=15);
```

## II. Box Plot

### `sns.boxplot()` function

```python
# 导入数据
fuel_econ = pd.read_csv('fuel-econ.csv')
fuel_econ.head()
```

<a id='ex1_box'></a>
### Example 1. Data Distribution in Efficiencies by Car Types Using `sns.boxplot()`

```python
# define a CategoricalDtype
sedan_classes = ['Minicompact Cars', 'Subcompact Cars', 'Compact Cars', 'Midsize Cars', 'Large Cars']

# transform the data type of 'VClass' volumn into a self-defined categorical type
vclasses = pd.api.types.CategoricalDtype(categories=sedan_classes, ordered=True)
fuel_econ['VClass'] = fuel_econ['VClass'].astype(vclasses)


base_color = sns.color_palette()[0]

# draw a box plot in the right axes
sns.boxplot(data=fuel_econ, x='VClass', y='comb', color=base_color)
plt.xticks(rotation=-20); 
```

```python
# define a CategoricalDtype
sedan_classes = ['Minicompact Cars', 'Subcompact Cars', 'Compact Cars', 'Midsize Cars', 'Large Cars']

# transform the data type of 'VClass' volumn into a self-defined categorical type
vclasses = pd.api.types.CategoricalDtype(categories=sedan_classes, ordered=True)
fuel_econ['VClass'] = fuel_econ['VClass'].astype(vclasses)

plt.figure(figsize=(16, 5))

base_color = sns.color_palette()[0]

# draw a violin plot in the left axes
plt.subplot(1,2,1)
ax1 = sns.violinplot(data=fuel_econ, x='VClass', y='comb', color=base_color, inner='quartile')
plt.xticks(rotation=-20);

# draw a box plot in the right axes
plt.subplot(1,2,2)
sns.boxplot(data=fuel_econ, x='VClass', y='comb', color=base_color)
plt.ylim(ax1.get_ylim()) # set y-axis limits to be same as left plot
plt.xticks(rotation=-20); 
```

___Comparisons Between Violin Plot and Box Plot___

Comparing the two plots, the `box plot` is a cleaner summary of the data than the `violin plot`. It's easier to compare statistics between the groups with a `box plot`. This makes a `box plot` worth more consideration if we have a lot of groups to compare, or if we are building explanatory plots. We can clearly see from the box plot that the `Large Cars` has the lowest median. 

On the other hand, the `box plot` lacks as nuanced a depiction of distributions as the `violin plot`: you can't see the slight bimodality present in the `Minicompact Cars`. The `violin plot` may be a better option for exploration, especially since `seaborn`'s implementation also includes the box plot by default.



**Variation: Horizontal Box Plot.**

```python
sns.boxplot(data=fuel_econ, x='comb', y='VClass', color=base_color);
```

### Example 2. Data Distribution in Displacement by Car Types Using `sns.boxplot()`

```python
def violin_displ_carTypes():
    """
    Solution for Question 1 in violin and box plot practice: plot the relationship
    between vehicle class and engine displacement.
    """
    sol_string = ["I used a violin plot to depict the data in this case; you might",
                  "have chosen a box plot instead. One of the interesting things",
                  "about the relationship between variables is that it isn't consistent.",
                  "Compact cars tend to have smaller engine sizes than the minicompact",
                  "and subcompact cars, even though those two vehicle sizes are smaller.",
                  "The box plot would make it easier to see that the median displacement",
                  "for the two smallest vehicle classes is greater than the third quartile",
                  "of the compact car class."]
    print((" ").join(sol_string))

    # data setup
    fuel_econ = pd.read_csv('fuel-econ.csv')

    sedan_classes = ['Minicompact Cars', 'Subcompact Cars', 'Compact Cars', 'Midsize Cars', 'Large Cars']
    pd_ver = pd.__version__.split(".")
    if (int(pd_ver[0]) > 0) or (int(pd_ver[1]) >= 21): # v0.21 or later
        vclasses = pd.api.types.CategoricalDtype(ordered = True, categories = sedan_classes)
        fuel_econ['VClass'] = fuel_econ['VClass'].astype(vclasses)
    
    else: # pre-v0.21
        fuel_econ['VClass'] = fuel_econ['VClass'].astype('category', ordered = True,
                                                         categories = sedan_classes)

    # plotting
    base_color = sns.color_palette()[0]
    sns.violinplot(data = fuel_econ, x = 'VClass', y = 'displ', color = base_color)
    plt.xticks(rotation = 15)
```

```python
violin_displ_carTypes()
```

___Conclusion:___

I used a violin plot to depict the data in this case; you might have chosen a box plot instead. One of the interesting things about the relationship between variables is that it isn't consistent. Compact cars tend to have smaller engine sizes than the minicompact and subcompact cars, even though those two vehicle sizes are smaller. The box plot would make it easier to see that the median displacement for the two smallest vehicle classes is greater than the third quartile of the compact car class.
